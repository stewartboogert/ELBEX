```python
import matplotlib.pyplot as _plt
import xsuite
import pybdsim

import sys
sys.path.append('../2_Ocelot')
import XTD8
import XTD8_1_shielding_moves

sys.path.append('../0_scripts')
import elbex_xsuite
import elbex_ocelot
```

### Convert OCELOT to XSuite

```python
xs = pybdsim.Convert.Ocelot2Xsuite(XTD8, line_name="XTD8")
xs.env.to_json('XTD8.json')
```

### Build trackers and calculate twiss

```python
xs.env.build_trackers()
twiss = xs.env.twiss(**xs.tws0)
```

### Plots

```python
elbex_xsuite.plotTwissXYAlongS(twiss,'XTD8','betx','bety',label=r"\beta")
elbex_xsuite.plotTwissXYAlongS(twiss,'XTD8','dx','dy',label=r"\eta")
```

```python
elbex_xsuite.writeXsuiteToExcel(xs,"XTD8.xlsx")
```

### Magnet moves to avoid shielding

```python
xs_1_shielding_moves = pybdsim.Convert.Ocelot2Xsuite(XTD8_1_shielding_moves, line_name="XTD8")
xs_1_shielding_moves.env.to_json('XTD8_1_shielding_moves.json')
xs_1_shielding_moves.env.build_trackers()
twiss_1_shielding_moves = xs_1_shielding_moves.env.twiss(**xs_1_shielding_moves.tws0)
elbex_xsuite.plotTwissXYAlongS(twiss_1_shielding_moves,'XTD8','betx','bety',label=r"\beta")
elbex_xsuite.plotTwissXYAlongS(twiss_1_shielding_moves,'XTD8','dx','dy',label=r"\eta")
elbex_xsuite.writeXsuiteToExcel(xs_1_shielding_moves,"XTD8_1_shielding_moves.xlsx")
```
