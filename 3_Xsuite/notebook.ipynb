```python
import matplotlib.pyplot as _plt
import xsuite
import pybdsim

import sys
sys.path.append('../2_Ocelot')
import XTD8
```

### Convert OCELOT to XSuite

```python
xs = pybdsim.Convert.Ocelot2Xsuite(XTD8, line_name="XTD8")
```

### Build trackers and calculate twiss

```python
xs.env.build_trackers()
twiss = xs.env.twiss(**xs.tws0)
```

### Quick plots

```python
_plt.subplot(3,1,1)
_plt.plot(twiss['XTD8']['s'], twiss['XTD8']['betx'])
_plt.plot(twiss['XTD8']['s'], twiss['XTD8']['bety'])
_plt.subplot(3,1,2)
_plt.plot(twiss['XTD8']['s'], twiss['XTD8']['dx'])
_plt.plot(twiss['XTD8']['s'], twiss['XTD8']['dy'])
```
