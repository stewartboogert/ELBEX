### Decode GDML for positions

```python
import pyg4ometry
import pandas
# from ipyvtklink.viewer import ViewInteractiveWidget
```

```python
r = pyg4ometry.gdml.Reader("./lxelgeomdump_d92ef317.gdml")
reg = r.getRegistry()
wl = reg.getWorldVolume()
```

### Draw entire geometry

```python
v  = pyg4ometry.visualisation.VtkViewerNew()
v.addLogicalVolume(wl)
v.buildPipelinesAppend()
# ViewInteractiveWidget(v.renWin,width=600, height=600)
v.view()
```

### Z ordered list of main LUXE components

```python
dd = {}
dd['daughterName'] = []
dd['daughterLogicalName'] = []
dd['x'] = []
dd['y'] = []
dd['z'] = []

for dv in wl.daughterVolumes :
    pos = dv.position.eval()
    dd['daughterName'].append(dv.name)
    dd['daughterLogicalName'].append(dv.logicalVolume.name)
    dd['x'].append(pos[0])
    dd['y'].append(pos[1])
    dd['z'].append(pos[2])

df = pandas.DataFrame(dd)
df_zsort = df.sort_values('z')
df_zsort[100:120]
```
